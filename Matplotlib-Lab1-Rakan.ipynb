```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
ch1 = [0,1,3,2,1,5,2,1,4,2,4,0]
ch2 = {"Name": ["Rakan","Rayan","Razan", "Rashed"],
       "Region": ["East", "East", "North", "South"],
        "Age" :  [20,30,40,50]}
ch2DF = pd.DataFrame(ch2, columns=["Name", "Region", "Age"])

plt.bar(ch1, ch1)
plt.xlabel("chart1")
plt.ylabel("chart1")
plt.title("chart1")
plt.show()

plt.bar(ch2DF["Name"],ch2DF["Region"])
plt.xlabel("Names")
plt.ylabel("Regions")
plt.title("chart2")
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
fruits = ['Apple', 'Orange', 'Pie', 'Apple Pie']
store = [53, 70, 22, 96]
plt.pie(store, labels=fruits,autopct="%1.1f%%", startangle=270, explode=[0.2,0.2,0.2,0.2])
plt.title("Pie Chart")
plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(fruits, store, marker='o' ,alpha=0.4, s=50)
plt.title("Scatter Chart")
plt.xlabel("Fruits")
plt.ylabel("Storage")
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(fruits, store, alpha=0.4, s=50)
plt.title("Scatter Chart", color='r', loc='left')
plt.xlabel("Fruits", fontsize=16)
plt.ylabel("Storage", fontsize=16)
plt.show()
```

## Q5: Create three line charts with Legends at left and colors.

```python
#plt.plot(range(1920,2011,10),[2,4,6,8,10], [2,3,5,7,11], color='b')
year = [1920,1930,1940,1950,1960,1970,1980,1990,2000,2010]
unemployment_rate=[9.1,3.1,12,6.9,7,5.5,6.2,8,5,4]
plt.plot(range(1920,2011,10),[6,3,4,7,8,6,5,6,8,3], c='blue')
plt.plot(range(1920,2011,10),[8,3,4,5,2,7,8,5,4,6], c='green')
plt.plot(range(1920,2011,10),[2,7,9,0,0,3,4,5,2,2], c='y')
plt.plot(year, unemployment_rate, c='r', linewidth=2)
plt.title("Unemployment Rate per Year")
plt.legend(['Blue','Green','Yellow','Red'], shadow=True, loc='upper left') # Sorry it can't be left
plt.xlabel("Year")
plt.ylabel("Unemployment Rate")
plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
plt.figure(figsize=[20,10])
plt.suptitle('Three charts')

plt.subplot(3,1,1)
plt.pie(store, labels=fruits,autopct="%1.1f%%");


plt.subplot(3,1,2)
plt.scatter(fruits, store, alpha=0.4, s=50);

plt.subplot(3,1,3)
plt.plot(year, unemployment_rate, c='r', linewidth=2);
```
